```python
from pybeamline.sinks.graphviz_sink import graphviz_sink
from pybeamline.mappers.dfg_str_to_graphviz import dfg_str_to_graphviz
from pybeamline.algorithms.discovery import simple_dfg_miner
from pybeamline.sources.string_test_source import string_test_source

string_test_source(["ABCD", "ACBD"]).pipe(
    simple_dfg_miner(model_update_frequency=1),
    dfg_str_to_graphviz()
).subscribe(graphviz_sink())
```
